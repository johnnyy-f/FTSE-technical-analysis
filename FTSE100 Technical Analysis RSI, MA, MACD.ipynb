```python
# Function to fetch historical daily prices for a given symbol using yfinance
def get_historical_data(symbol):
    try:
        stock_data = yf.download(symbol, period='1Y')  # Adjust the date range
        return stock_data
    except Exception as e:
        print(f"Error fetching historical data for {symbol}: {e}")
        return None

# Function to calculate RSI using talib
def calculate_rsi(data, period=14):
    try:
        rsi = talib.RSI(data['Close'], timeperiod=period)
        return rsi.iloc[-1]  # Get the last RSI value
    except Exception as e:
        print(f"Error calculating RSI: {e}")
        return None

# Create a DataFrame to store RSI values
rsi_data = pd.DataFrame(index=tickers, columns=['RSI'])

# Fetch historical data and calculate RSI for each UK stock
for ticker in tickers:
    stock_data = get_historical_data(ticker)
    if stock_data is not None:
        rsi_value = calculate_rsi(stock_data)
        rsi_data.loc[ticker, 'RSI'] = rsi_value

# Display the DataFrame with RSI values
print(rsi_data)

#Export RSI Data to a CSV file
rsi_data.to_csv(r'#YOURFILEPATH/rsi_data.csv')
```

```python
# List of FTSE 100 tickers
ftse100_tickers = tickers

# Function to fetch most recent daily prices for a given symbol using yfinance
def get_stock_data(symbol):
    try:
        stock_data = yf.download(symbol, period='1Y')
        return stock_data
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return None

# Function to calculate moving average difference and filter by proximity to zero
def calculate_ma_difference(data, short_window=50, long_window=200, threshold=0.01):
    try:
        # Calculate short-term and long-term moving averages
        data['Short_MA'] = data['Close'].rolling(window=short_window, min_periods=50).mean()
        data['Long_MA'] = data['Close'].rolling(window=long_window, min_periods=200).mean()

        # Calculate the difference between short-term and long-term moving averages
        data['MA_Difference'] = abs(data['Short_MA'] - data['Long_MA'])

        return data[['Close', 'Short_MA', 'Long_MA', 'MA_Difference']].iloc[-1]  # Select the last row

    except Exception as e:
        print(f"Error calculating moving average difference: {e}")
        return None

# Create an empty DataFrame to store the most recent data for each stock
result_df = pd.DataFrame(columns=['Ticker', 'Close', 'Short_MA', 'Long_MA', 'MA_Difference'])

# Fetch most recent data and calculate moving average difference for each FTSE 100 stock
for ticker in ftse100_tickers:
    stock_data = get_stock_data(ticker)
    if stock_data is not None:
        ma_difference_data = calculate_ma_difference(stock_data)
        result_df = pd.concat([result_df, pd.DataFrame({'Ticker': [ticker], **ma_difference_data.to_dict()})], ignore_index=True)

# Display the DataFrame with the most recent data for each stock
print(result_df)

#Export Daily Moving Average Data to a CSV file
result_df.to_csv(r'#YOURFILEPATH/dailyma.csv')
```

```python
# List of FTSE 100 tickers
ftse100_tickers = tickers

# Function to fetch most recent daily prices for a given symbol using yfinance
def get_stock_data(symbol):
    try:
        stock_data = yf.download(symbol, period='1Y')
        return stock_data
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return None

# Function to calculate MACD and filter by proximity to zero
def calculate_macd(data, short_window=12, long_window=26, signal_window=9, threshold=0.01):
    try:
        # Calculate short-term and long-term exponential moving averages
        data['Short_EMA'] = data['Close'].ewm(span=short_window, adjust=False).mean()
        data['Long_EMA'] = data['Close'].ewm(span=long_window, adjust=False).mean()

        # Calculate MACD line
        data['MACD'] = data['Short_EMA'] - data['Long_EMA']

        # Calculate signal line
        data['Signal_Line'] = data['MACD'].ewm(span=signal_window, adjust=False).mean()

        # Calculate the difference between MACD and Signal Line
        data['MACD_Signal_Difference'] = abs(data['MACD'] - data['Signal_Line'])

        return data[['Close', 'Short_EMA', 'Long_EMA', 'MACD', 'Signal_Line', 'MACD_Signal_Difference']].iloc[-1]  # Select the last row

    except Exception as e:
        print(f"Error calculating MACD: {e}")
        return None

# Create an empty DataFrame to store the most recent data for each stock
result_df_macd = pd.DataFrame(columns=['Ticker', 'Close', 'Short_EMA', 'Long_EMA', 'MACD', 'Signal_Line', 'MACD_Signal_Difference'])

# Fetch most recent data and calculate MACD for each FTSE 100 stock
for ticker in ftse100_tickers:
    stock_data = get_stock_data(ticker)
    if stock_data is not None:
        macd_data = calculate_macd(stock_data)
        result_df_macd = pd.concat([result_df_macd, pd.DataFrame({'Ticker': [ticker], **macd_data.to_dict()})], ignore_index=True)

# Display the DataFrame with the most recent MACD data for each stock
print(result_df_macd)

#Export MACD Data to a CSV file
result_df_macd.to_csv(r'#YOURFILEPATH/result_df_macd.csv')
```
